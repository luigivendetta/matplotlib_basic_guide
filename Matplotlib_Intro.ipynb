# Customizing Our Plot:


`making it look like our .png file`

```python
plt.style.use('default')
fig, ax = plt.subplots(figsize = (15,5))
over50.plot(kind="scatter",
           x = 'age',
           y = 'chol',
           c = 'target',
           ax = ax);
ax.axhline(over50["chol"].mean(), linestyle='dotted')
#ax.plot(list(range(len(mean))), mean)
#plotting Data
scatter = ax.scatter(x = over50["age"],
                    y = over50["chol"],
                    c = over50["target"],)
##customizing figure
ax.set(title="Heart Disease and Cholesterol Levels ",
      xlabel = "Ages",
      ylabel = "Cholesterol ")
##adding a legend
ax.legend(*scatter.legend_elements(),
           title ="Target "  );
fig.savefig("Heart_Disease_and_Cholesterol_Levels.png")
```

```python
fig, ax = plt.subplots(figsize = (15,10), nrows=2,ncols=1)
#plotting Data
scatter = ax[0].scatter(x = over50["age"],
                    y = over50["chol"],
                    c = over50["target"],)
ax[0].axhline(over50["chol"].mean(), linestyle='dotted')
scatter2 = ax[1].scatter(x= over50["age"],
                        y= over50["thalach"],
                        c= over50["target"])
ax[1].axhline(over50["thalach"].mean(), linestyle="dotted")
##customizing figure
ax[0].set(title="Heart Disease and Cholesterol Levels ",
      xlabel = "Ages",
      ylabel = "Cholesterol ")
ax[1].set(title="Heart Disease and Heighest Heart-Rate Levels ",
      xlabel = "Ages",
      ylabel = "Heart-Rate ")

##adding a legend
ax[0].legend(*scatter.legend_elements(),
           title ="Target "  );
ax[1].legend(*scatter2.legend_elements(),
            title = "Target");
#fig.savefig("Heart_Disease_and_Cholesterol_Levels_and_Hear_Rate2.png")
```

```python
fig, ax = plt.subplots(figsize = (15,10), nrows=2,ncols=1, sharex= True)
#plotting Data
scatter = ax[0].scatter(x = over50["age"],
                    y = over50["chol"],
                    c = over50["target"],)
ax[0].axhline(over50["chol"].mean(), linestyle='dotted')
scatter2 = ax[1].scatter(x= over50["age"],
                        y= over50["thalach"],
                        c= over50["target"])
ax[1].axhline(over50["thalach"].mean(), linestyle="dotted")
##customizing figure
ax[0].set(title="Heart Disease and Cholesterol Levels ",
     # xlabel = "Ages",
      ylabel = "Cholesterol ")
ax[1].set(title="Heart Disease and Heighest Heart-Rate Levels ",
      xlabel = "Ages",
      ylabel = "Heart-Rate ")

##adding a legend
ax[0].legend(*scatter.legend_elements(),
           title ="Target "  );
ax[1].legend(*scatter2.legend_elements(),
            title = "Target");
fig.suptitle("Heart Disease Analysis ", fontsize=14, fontweight = "bold");
fig.savefig("Heart_Disease_Analysis_with_no_customization.png")
```


## x and y axis limit


```python
plt.style.use("seaborn-whitegrid")
fig, ax = plt.subplots(figsize = (15,10), nrows=2,ncols=1, sharex= True)
#plotting Data
scatter = ax[0].scatter(x = over50["age"],
                    y = over50["chol"],
                    c = over50["target"],
                       cmap="winter")
ax[0].axhline(over50["chol"].mean(), linestyle='dotted')
scatter2 = ax[1].scatter(x= over50["age"],
                        y= over50["thalach"],
                        c= over50["target"],
                        cmap="winter")
ax[1].axhline(over50["thalach"].mean(), linestyle="dotted")
##customizing figure
ax[0].set(title="Heart Disease and Cholesterol Levels ",
     # xlabel = "Ages",
      ylabel = "Cholesterol ")
ax[0].set_xlim([50,80])
ax[0].set_ylim([100,600])
ax[1].set(title="Heart Disease and Heighest Heart-Rate Levels ",
      xlabel = "Ages",
      ylabel = "Heart-Rate ")
ax[1].set_ylim([60,200])
##adding a legend
ax[0].legend(*scatter.legend_elements(),
           title ="Target "  );
ax[1].legend(*scatter2.legend_elements(),
            title = "Target");
fig.suptitle("Heart Disease Analysis ", fontsize=14, fontweight = "bold");
#fig.savefig("Heart_Disease_Analysis.png")
```
